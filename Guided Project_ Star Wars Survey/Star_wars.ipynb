### Convert Yes/No to True/False for easier selection

```python
yes_no = {
    "Yes": True,
    "No": False
}
cols=['Have you seen any of the 6 films in the Star Wars franchise?','Do you consider yourself to be a fan of the Star Wars film franchise?']

for col in cols:
    star_wars[col]=star_wars[col].map(yes_no)
```

### Convert the checkboxes with movie names to True/False

```python
import numpy as np
movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_mapping)
```

```python
star_wars = star_wars.rename(columns={
        "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"
        })

star_wars.head()
```

```python
star_wars.columns[9:15]
```

```python

for col in star_wars.columns[9:15]:
    star_wars[col] = star_wars[col].astype(float)
    
star_wars = star_wars.rename(columns={
        "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
        "Unnamed: 10": "ranking_2",
        "Unnamed: 11": "ranking_3",
        "Unnamed: 12": "ranking_4",
        "Unnamed: 13": "ranking_5",
        "Unnamed: 14": "ranking_6"
        })

star_wars.head()
```

### Mean ranking

```python
mean_rank=star_wars[['ranking_1','ranking_2','ranking_3','ranking_4','ranking_5','ranking_6']].mean()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.bar(range(6),mean_rank)
```

```python
mean_rank.plot.bar()
```

We selected the existing respond_ID, map Yes/No to True/False in the fan/seen coloumns, and converted the ranking columns to float type (so that we can calculate the mean afterward). Also, we renamed some columns for easier inspection.

The fifth starwars V The Empire Strikes Back seems to be ranked among the highest for most of responders. Also, the original three are much more popular than the new ones.

The V and VI were seen by most people, while the new-three have decreasing popularity. Again, the V was seen by the most people.

### Star Wars fans

```python
swfan = star_wars[star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] == True]
no_swfan = star_wars[star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] == False]
```

```python
mean_fan=swfan[['ranking_1','ranking_2','ranking_3','ranking_4','ranking_5','ranking_6']].mean()
mean_nofan=no_swfan[['ranking_1','ranking_2','ranking_3','ranking_4','ranking_5','ranking_6']].mean()

mean_fan.plot.bar()
mean_nofan.plot.bar(color='r',width=0.2)
plt.legend(['fan','no_fan'])
```

```python
swfan.filter(regex=("^seen_?")).sum().plot.bar()
no_swfan.filter(regex=("^seen_?")).sum().plot.bar(color='r',width=0.2)
```
